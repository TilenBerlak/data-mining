```python
%matplotlib inline
import matplotlib
%config InlineBackend.figure_format = 'jpg'
matplotlib.figure.Figure.__repr__ = lambda self: (
    f"<{self.__class__.__name__} size {self.bbox.size[0]:g}"
    f"x{self.bbox.size[1]:g} with {len(self.axes)} Axes>")
import matplotlib.pyplot as plt
plt.style.use('PR.mplstyle')
```

```python
apl_price = [93.95, 112.15, 104.05, 144.85, 169.49]
ms_price = [39.01, 50.29, 57.05, 69.98, 94.39]
year = [2014, 2015, 2016, 2017, 2018]
```

```python
plt.plot(year, apl_price,
        year, ms_price)
plt.axis([2013, 2019, 20, 180])
plt.xlabel("Year")
plt.ylabel("Stock price")
plt.show()
```

```python
fig_1 = plt.figure(1, figsize=(5, 4.8))

chart_1 = fig_1.add_subplot(211)
chart_2 = fig_1.add_subplot(212)

chart_1.plot(year, apl_price)
chart_2.scatter(year, ms_price)
```
